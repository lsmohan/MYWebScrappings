```python
f_string = open('Kusakratha_VishnuSahasraNama.txt', 'r', encoding="utf-8").read()
kxml = open('kusakratha-vishnu.xml', 'w', encoding="utf-8")
```

```python
splits = f_string.split("Text")
```
