```python
from bs4 import BeautifulSoup

files = ['Slokha 01-12.html', 'Slokha 13-24.html']

f = open('vishnu-trisiyengar.txt', 'w')

for file in files:
    html = open(file, 'r').read()
    soup = BeautifulSoup(html, "lxml")
    
    tabs = soup.findAll('table')
    vishnu_sahas = tabs[4]
    rows = vishnu_sahas.findAll('tr')
    shloka = rows[3]
    names = shloka.findAll('p')
    
    for n in names:
        if (n.get('style') == "margin-left: 25; margin-top: 0px; margin-bottom: 0px"):
            f.write(str(n.text))
            f.write('\n')
        elif(n.get('style') == "MARGIN-TOP: 0px; MARGIN-BOTTOM: 0px"):
            f.write(str(n.text))
            f.write('\n')
        elif(n.get('style') == "word-spacing: 0; line-height: 100%; margin-left: 10; margin-right: 10; margin-top: 0; margin-bottom: 0"):
            f.write(str(n.text))
            f.write('\n')
    
```
